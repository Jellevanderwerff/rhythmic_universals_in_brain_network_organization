```python
import pandas as pd
import thebeat
import matplotlib.pyplot as plt
import os
import numpy as np
```

```python
# Load ITIs df
df = pd.read_csv(os.path.join('data', 'experiment', 'processed', 'ITIs.csv'))
```

```python
plt.style.use('seaborn-v0_8')

fig, axs = plt.subplots(30, 3, figsize=(15, 90), tight_layout=True, dpi=300)

sequence_ids = np.random.choice(df.sequence_id.unique(), 90)
df_short = df[df.sequence_id.isin(sequence_ids)]

count = 0

for seq_id, seq_df in df_short.groupby('sequence_id'):
    stimulus = thebeat.Sequence(seq_df['stim_ioi'].values)
    response = thebeat.Sequence(seq_df['resp_iti'].values)
    response_norm = thebeat.Sequence(seq_df['resp_iti_norm'].values)
    response_norm_q = thebeat.Sequence(seq_df['resp_iti_norm_q'].values)
    stimulus_q = thebeat.Sequence(seq_df['stim_ioi_q'].values)


    thebeat.visualization.plot_multiple_sequences([stimulus, response, response_norm, response_norm_q, stimulus_q], 
                                                  y_axis_labels=['Stimulus\n(original)', 
                                                                 'Response\n(original)', 'Response\n(normalized)', 
                                                                 'Response\n(normalized+quantized)',
                                                                 'Stimulus\n(quantized)'],
                                                   ax=axs[count // 3, count % 3])
    count += 1
```
