```python
import pandas as pd
import thebeat
import os
import matplotlib.pyplot as plt
```

```python
# Load ITIs df
ITIs = pd.read_csv(os.path.join('data', 'experiment', 'processed', 'ITIs.csv'))
ITIs_bytrial = pd.read_csv(os.path.join('data', 'experiment', 'processed', 'ITIs_bytrial.csv'))
pp_measures = pd.read_csv(os.path.join('data', 'experiment', 'processed', 'pp_measures.csv'))
```

```python
# Find sequence id's of trials with highest and lowest edit distance
highest_asynchrony = ITIs_bytrial.nlargest(18, 'asynchrony_norm_abs_trialsum').sequence_id.values
lowest_asynchrony = ITIs_bytrial.nsmallest(18, 'asynchrony_norm_abs_trialsum').sequence_id.values

# Find sequence id's of trials with middle edit distance
middle_async_a = len(ITIs_bytrial) // 2 - 9
middle_async_b = len(ITIs_bytrial) // 2 + 9
middle_async = ITIs_bytrial.sort_values(by='asynchrony_norm_abs_trialsum').iloc[middle_async_a:middle_async_b].sequence_id.values

# Get ITIs of 9 sequences with the highest and lowest edit distance
highest_async_df = ITIs[ITIs['sequence_id'].isin(highest_asynchrony)]
middle_async_df = ITIs[ITIs['sequence_id'].isin(middle_async)]
lowest_async_df = ITIs[ITIs['sequence_id'].isin(lowest_asynchrony)]
```

# Lowest asynchrony

```python
# Plot ITIs by sequence_id
plt.style.use('seaborn-v0_8')

fig, axs = plt.subplots(6, 3, figsize=(15, 20), tight_layout=True, dpi=300)

count = 0

for seq_id, seq_df in lowest_async_df.groupby('sequence_id'):
    resp_seq = thebeat.Sequence(seq_df.resp_iti_norm.values)
    stim_seq = thebeat.Sequence(seq_df.stim_ioi.values)
    thebeat.visualization.plot_multiple_sequences([stim_seq, resp_seq], y_axis_labels=['Stimulus\n(original)', 'Response\n(tempo normalized)'],
                                                  ax = axs.flat[count],
    title=
f"Entropy difference = {seq_df.entropy_diff.values[0]:.2f}\n \
Asynchrony = {ITIs_bytrial[ITIs_bytrial.sequence_id == seq_id].asynchrony_norm_abs_trialsum.values[0]:.4f}\n \
Simple ratio increase = {seq_df.simple_ratio_increase.values[0]}\n \
G = {seq_df.G_resp.values[0]:.2f}\n \
Sequence id = {seq_id}")

    count += 1

# Change font size of title in all subplots
for ax in axs.flat:
    ax.title.set_fontsize(10)
```
