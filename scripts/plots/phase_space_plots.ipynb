# NPOR: Plots

```python
import thebeat
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
# Import df
df = pd.read_csv(os.path.join("data", "experiment", "processed", "ITIs.csv"))
```

# Phase space plots

```python
# Let's try one first

# set plt style to seaborn
plt.style.use('seaborn-v0_8')

fig, axs = plt.subplots(16, 6, tight_layout=True, figsize=(10, 20))

count = 0

for pp_id, pp_df in df.groupby('pp_id_behav'):
    for seq_id, seq_df in pp_df.groupby('sequence_id'):
        stim_seq = thebeat.Sequence(seq_df['stim_ioi'].values)
        resp_seq = thebeat.Sequence(seq_df['resp_iti'].values)
        thebeat.visualization.phase_space_plot(stim_seq, ax=axs.flat[count])
        thebeat.visualization.phase_space_plot(resp_seq, ax=axs.flat[count + 1])
    count += 2

for ax in axs.flat:
    ax.set_xticks([])
    ax.set_yticks([])
    ax.set_xlabel('')
    ax.set_ylabel('')
    ax.set_aspect('equal')
```
